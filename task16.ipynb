```python
# Q1. Load the insurance dataset
# Display the first 10 rows

import pandas as pd

df = pd.read_csv("insurance.csv")

df.head(10)
```

```python
# Q2(a). Display the shape of the dataset
df.shape
```

```python
# Q2(b). Display column names and data types
df.info()
```

```python
# Q2(c). Display statistical summary
df.describe()
```

```python
# Q3. Check null or missing values
df.isnull().sum()
```

```python
# Q4(a). Display numerical columns

df.select_dtypes(include=["int64", "float64"]).columns
```

```python
# Q4(b). Display categorical columns
df.select_dtypes(include=["object"]).columns
```

```python
# Q5(a). Distribution plot of Age

import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df[" age"], kde=True)

plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
# Q5(b). Distribution plot of BMI
sns.histplot(df["bmi"], kde=True)
plt.title("Distribution of BMI")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
# Q5(c). Distribution plot of Children
sns.histplot(df["children"], kde=True)

plt.title("Distribution of Children")
plt.xlabel("Children")
plt.ylabel("Frequency")

plt.show()
```

```python
# Q5(d). Distribution plot of Expenses
sns.histplot(df["expenses"], kde=True)
plt.title("Distribution of Expenses")
plt.xlabel("Expenses")
plt.ylabel("Frequency")
plt.show()
```

```python
# Q6(a). Count plot of Sex
sns.countplot(x="sex", data=df)
plt.title("Count Plot of Sex")
plt.show()
```

```python
# Q6(b). Count plot of Smoker
sns.countplot(x="smoker", data=df)
plt.title("Count Plot of Smoker")
plt.show()
```

```python
# Q6(c). Count plot of Region
sns.countplot(x="region", data=df)
plt.title("Count Plot of Region")
plt.show()
```

```python
# Q7. Correlation Heatmap
correlation = df.corr(numeric_only=True)
plt.figure(figsize=(7,5))
sns.heatmap(correlation,
            annot=True,
            cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
# Q8(a). Average insurance expenses
df["expenses"].mean()
```

```python
# Q8(b). Maximum insurance expenses
df["expenses"].max()
```

```python
# Q8(c). Minimum insurance expenses
df["expenses"].min()
```

```python
# Q8(d). Compare average expenses
# of smokers and non-smokers
df.groupby("smoker")["expenses"].mean()
```

```python
# Q9(a). Boxplot of Expenses by Smoker
sns.boxplot(x="smoker",
            y="expenses",
            data=df)
plt.title("Expenses by Smoker")
plt.show()
```

```python
# Q9(b). Boxplot of Expenses by Sex

sns.boxplot(x="sex",
            y="expenses",
            data=df)

plt.title("Expenses by Sex")

plt.show()
```
